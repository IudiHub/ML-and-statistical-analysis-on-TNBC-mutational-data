```python
# sostituisci i missing values nella colonna ENSP con la parola 'VUOTO'
def sostituzione_nan (df_relevant_genes_HS_ENSP):

    for index, row in df_relevant_genes_HS_ENSP.iterrows():
        gene = row.Hugo_Symbol
        ENSP =  str(row.ENSP)
    #print(type(ENSP))
        if ENSP.startswith('ENSP') is False:
            df_relevant_genes_HS_ENSP.loc[index, 'ENSP'] = 'VUOTO' # se non c'è l'ENSP della proteina metti 'VUOTO'
            #count_nan = count_nan+1
    return (df_relevant_genes_HS_ENSP)
```

```python
# substitute the ENSP NaN with 'VUOTO'
data_mutation_SINIFICANTgenes_onlyTN_P005_gene_ENSP_nonan = sostituzione_nan(data_mutation_SINIFICANTgenes_onlyTN_P005_gene_ENSP)
data_mutation_SINIFICANTgenes_onlyTN_P005_gene_ENSP_nonan
```

```python
## The function Vett_transcript takes all the unique Hugo Symbols as indexes and create the column 'Protein length' filled with zeros 
def Index_relevant_genes (df_mut):
    vett_Hugo_symbols_relevant = df_mut.Hugo_Symbol.unique()
    df_vett_Hugo_symbols_relevant = pd.DataFrame(index = vett_Hugo_symbols_relevant, columns = ['Protein_length'])
    df_vett_Hugo_symbols_relevant = df_vett_Hugo_symbols_relevant.fillna(0)
    return (df_vett_Hugo_symbols_relevant)
```

```python
Hugo_symbols_relevant_SIGNIFICANT_genes_onlyTN_P005 = Index_relevant_genes(data_mutation_SINIFICANTgenes_onlyTN_P005_gene_ENSP_nonan)
Hugo_symbols_relevant_SIGNIFICANT_genes_onlyTN_P005
```

### Retrieve the protein length using pyensembl

```python
from pyensembl import EnsemblRelease

data = EnsemblRelease(75)

def protein_length_computation(vett_relevant_genes, df_relevant_genes_HS_ENSP_nonan):
    
    for index, row in vett_relevant_genes.iterrows():
        gene = index
        length = row.Protein_length
        if length == 0:
            df = df_relevant_genes_HS_ENSP_nonan[df_relevant_genes_HS_ENSP_nonan['Hugo_Symbol'] == gene]
            for index, row in df.iterrows():
                ENSP_singolo = row.ENSP
                if ENSP_singolo != 'VUOTO':
                #print(ENSP_singolo)
                    protein = data.protein_sequence(ENSP_singolo)
                    protein_length = len(protein)
                    vett_relevant_genes.loc[gene, 'Protein_length'] = protein_length
```

```python
# Define a function that takes in input the mutation dataset containing only the relevant genes and gives in output a new df containing HugoSymbol e Protein Position
def protein_position_Hugo_Symbol_only(relevant_mutations):
    df_positions = []

    for index, row in relevant_mutations.iterrows():
        df_positions.append([row.Hugo_Symbol, row.Protein_position_start])
    
    df_positions = pd.DataFrame(df_positions, columns = ['symbol', 'position'])
    df_positions = df_positions.dropna() # drop the protein position missing values 
    
    # Convert single column to int dtype
    df_positions['position'] = pd.to_numeric(df_positions['position'])
    df_positions = df_positions.astype({"position":"int"})
    
    return (df_positions)
```

```python
# Define a function that gives in output a new df containing HugoSymbol e Protein Length
def protein_length_only_Hugo_Symbol(relevant_mutations):  
    df_protein_lengths = []
    

    for index, row in relevant_mutations.iterrows():
        df_protein_lengths.append([row.Hugo_Symbol, row.Protein_length])
    
    df_protein_lengths = pd.DataFrame(df_protein_lengths, columns = ['symbol', 'length'])
    df_protein_lengths = df_protein_lengths.dropna()
    df_protein_lengths =  df_protein_lengths.drop_duplicates()
    
    return (df_protein_lengths)
```
