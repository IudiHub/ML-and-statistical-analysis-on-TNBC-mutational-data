```python
# select all the rows containing the ENSP protein ID
genes_ENSP_selected=genes_ENSP[genes_ENSP.apply(lambda row: row.astype(str).str.startswith('ENSP').any(), axis=1)].copy()
genes_ENSP_selected
```

```python
# Function to extract 'ENSP' values from a row
def extract_ensp_values(row):
    ensp_values = []
    for value in row.values:
        if isinstance(value, str) and value.startswith('ENSP'):     #isinstance() to check if 'value' is a string
            ensp_values.append(value)
    return ensp_values

# Create a new column with 'ENSP' values for each row

genes_ENSP_selected['ENSP'] = genes_ENSP_selected.apply(lambda row: extract_ensp_values(row), axis=1)

# Print the resulting DataFrame
genes_ENSP_selected
```

```python
Hugo_symbols_selected=genes_ENSP_selected[['Hugo_Symbol', 'Tumor_Sample_Barcode', 'Chromosome', 'Start_Position', 'End_Position']]
Hugo_symbols_selected
```

```python
x=genes_ENSP_selected['ENSP'].str.get(0).to_frame()
x
```

```python
genes_ENSP_final = pd.concat([Hugo_symbols_selected, x], axis=1,ignore_index=False )
genes_ENSP_final.rename(columns={0: "Hugo_Symbol"}, inplace=True)
genes_ENSP_final.rename(columns={1: "ENSP"}, inplace=True)
genes_ENSP_final
```
